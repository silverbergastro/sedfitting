```python
import pandas as pd
from scipy import interpolate
import numpy as np
import matplotlib.pyplot as plt

spline_df = pd.read_csv('cont_paper_table_data_tycho_spline_input.csv')
spline_data = spline_df.values

spline_BTminVT = spline_data[:,0]

spline_VminVT = spline_data[:,1]

spline_BminV = spline_data[:,3]

tck_VminVT = interpolate.splrep(spline_BTminVT, spline_VminVT, s=0)
tck_BminV = interpolate.splrep(spline_BTminVT, spline_BminV, s=0)

input_target_df = pd.read_csv('cont_paper_table_data_tycho_target_input.csv')
input_target_data = input_target_df.values

num_targets = input_target_data[:,0].size

for i in range(num_targets):
    #print input_target_data[i,0], input_target_data[i,2]
    
    btmag = input_target_data[i,4]
    bterr = input_target_data[i,5]
    vtmag = input_target_data[i,6]
    vterr = input_target_data[i,7]
    
    btminvt = btmag - vtmag
    
    btminvt_err = np.sqrt((bterr**2)+(vterr**2))
    
    btminvt_ens = np.random.normal(btminvt, btminvt_err, 10000)
    
    #plt.hist(btminvt_ens)
    #plt.show()
    
    vminvt_ens = interpolate.splev(btminvt_ens, tck_VminVT, der=0)
    
    #plt.hist(vminvt_ens)
    #plt.show()
    
    vminvt_avg = np.mean(vminvt_ens)
    vminvt_std = np.std(vminvt_ens)
    
    bminv_ens = interpolate.splev(btminvt_ens, tck_BminV, der=0)
    
    #plt.hist(bminv_ens)
    #plt.show()
    
    bminv_avg = np.mean(bminv_ens)
    bminv_std = np.std(bminv_ens)
    
    vmag = vtmag + vminvt_avg
    verr = np.sqrt((vterr**2)+(vminvt_std**2))
    
    bmag = vmag + bminv_avg
    berr = np.sqrt((verr**2)+(bminv_std**2))
    
    print input_target_data[i,0], input_target_data[i,2], "BTmag = %.3f +/- %.3f, VTmag = %.3f +/- %.3f, BT-VT = %.3f +/- %.3f, V-VT = %.3f +/- %.3f, B-V = %.3f +/- %.3f, Bmag = %.3f +/- %.3f, Vmag = %.3f +/- %.3f" % (btmag, bterr, vtmag, vterr, btminvt, btminvt_err, vminvt_avg, vminvt_std, bminv_avg, bminv_std, bmag, berr, vmag, verr)

    #print "BT-VT = %.3f +/- %.3f, V-VT = %.3f +/- %.3f, B-V = %.3f +/- %.3f" % ()
    
    #print "Bmag = %.3f +/- %.3f, Vmag = %.3f +/- %.3f" % ()
```
